```python
import json
from collections import defaultdict

with open('../data/weekly_earnings.json', 'r') as f:
    data = json.load(f)
    columns = data['meta']['view']['columns']
    result = dict()
    for col in columns[8:]:
        result[col['name']] = []
    
    earnings = data['data']
    dicts = []
    for row in earnings:
        new_dict = defaultdict(lambda: [])
        for i, col in enumerate(columns[8:]):
            new_dict[col['name']].append(row[i+8])
        dicts.append(dict(new_dict))
    print(dicts)

    json.dump(dicts, open('../data/weekly_earnings_clean.json', 'w'),indent = 4)
```
